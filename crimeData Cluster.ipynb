# Kmeans clustering

```python
import pandas as pd
import matplotlib.pylab as plt
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist 
import numpy as np
```

```python
# Normalized data frame (considering the numerical part of data)
def_norm=norm_func(data.iloc[:,1:])
def_norm
```

```python
###### screw plot or elbow curve ############
k = list(range(2,15))
k
```

```python
TWSS = [] # variable for storing total within sum of squares for each kmeans 
```

```python
for i in k:
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(def_norm)
    WSS = [] # variable for storing within sum of squares for each cluster 
    for j in range(i):
        WSS.append(sum(cdist(def_norm.iloc[kmeans.labels_==j,:],kmeans.cluster_centers_[j].reshape(1,def_norm.shape[1]),"euclidean")))
    TWSS.append(sum(WSS))
```

```python
# Scree plot 
plt.plot(k,TWSS, 'ro-');plt.xlabel("No_of_Clusters");plt.ylabel("total_within_SS");plt.xticks(k)
```

```python
# Selecting 5 clusters from the above scree plot which is the optimum number of clusters 
model=KMeans(n_clusters=5).fit(def_norm)
```

```python
model.labels_ # getting the labels of clusters assigned to each row 
md=pd.Series(model.labels_)  # converting numpy array into pandas series object 
data['clust']=md # creating a  new column and assigning it to new column 
data.head()
```

# DBSCAN Clustering

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
data2=data.iloc[:,1:]
data2.head()
```

```python
array=data2.values
array
```

```python
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array)
X
```

```python
dbscan = DBSCAN(eps=0.8, min_samples=6)
dbscan.fit(X)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
cl
```

```python
pd.concat([data,cl],axis=1)
```
